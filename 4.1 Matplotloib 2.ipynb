```python
# import necesssary packages
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
x = [4,5,2,5,7,9,5,8,3,6,7,4,5,7,6,9,9,8,5,5,4,7]
```

```python
pd.Series(x).value_counts()
```

```python
plt.hist(x, bins=8)
```

Histogram shows frequency of given values on y-axis while Bar plot needs two values (one is for x and corresponding one is for y)


```python
sns.distplot(x, bins=len(x))
```

# Formatting Plots

```python
x = [0,1,2,3,4]
y = [5,7,4,6,7]
```

```python
plt.bar(x, y, width=[0.2,0.5,0.3,0.7,1], color=['red', 'blue', 'yellow'])
```

```python
plt.plot(x,y, lw=3, ls=':', marker='o',mew=15, color='red')
```

# Labels and Annotations

```python
plt.plot(x,y)
```

# SHIFT + TAB for documentation

```python
plt.xlabel('Age', fontsize=18, color='red')
plt.ylabel('Height', fontsize=22, color='green')
plt.title('Age vs Height of Prople', color='blue')
plt.plot(x,y)
```
